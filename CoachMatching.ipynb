```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set up display options
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)

# Load the datasets
coaches = pd.read_csv("C:/Users/hamou/Downloads/data/male_coaches.csv")
teams = pd.read_csv("C:/Users/hamou/Downloads/data/male_teams.csv")

# Display basic info
print("Coaches dataset shape:", coaches.shape)
print("Teams dataset shape:", teams.shape)
```

```python
# Explore coaches data
print("\nCoaches data columns:")
print(coaches.columns.tolist())
print("\nMissing values in coaches data:")
print(coaches.isnull().sum())

# Explore teams data
print("\nTeams data columns:")
print(teams.columns.tolist())
print("\nMissing values in teams data:")
print(teams.isnull().sum())

# Clean coaches data
# Convert dob to age
current_year = datetime.now().year
coaches['age'] = pd.to_datetime(coaches['dob']).dt.year.apply(lambda x: current_year - x)

# Drop columns with high missing values or less relevance
coaches_clean = coaches.drop(['face_url', 'coach_url'], axis=1)

# Clean teams data
# Identify columns with all null values
null_columns = teams.columns[teams.isnull().all()]
teams_clean = teams.drop(null_columns, axis=1)

# Keep only relevant columns for our analysis
relevant_team_columns = [
    'team_id', 'team_name', 'league_name', 'league_level', 
    'nationality_name', 'overall', 'attack', 'midfield', 
    'defence', 'coach_id', 'international_prestige', 
    'domestic_prestige', 'starting_xi_average_age', 
    'whole_team_average_age'
]
teams_clean = teams_clean[relevant_team_columns]

# Display cleaned data
print("\nCleaned coaches data shape:", coaches_clean.shape)
print("Cleaned teams data shape:", teams_clean.shape)
```

```python
# Merge datasets on coach_id
merged_data = pd.merge(teams_clean, coaches_clean, on='coach_id', how='left')

# Calculate some new features
# Coach's experience (simplified as age for now)
merged_data['coach_experience'] = merged_data['age']

# Team quality metrics
merged_data['team_quality_ratio'] = merged_data['overall'] / (merged_data['attack'] + merged_data['midfield'] + merged_data['defence'])

# Nationality alignment (whether coach and team share nationality)
merged_data['nationality_alignment'] = (merged_data['nationality_name_x'] == merged_data['nationality_name_y']).astype(int)

# Display merged data
print("\nMerged data shape:", merged_data.shape)
print("Merged data columns:", merged_data.columns.tolist())
```

```python
# For this exercise, we'll define success as:
# 1. Team overall rating (continuous)
# 2. Top tier classification (binary - above/below median)

median_overall = merged_data['overall'].median()
merged_data['success_top_tier'] = (merged_data['overall'] >= median_overall).astype(int)

print("\nSuccess metric distribution:")
print(merged_data['success_top_tier'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Select features and target
features = [
    'coach_experience', 'nationality_alignment',
    'international_prestige', 'domestic_prestige',
    'starting_xi_average_age', 'whole_team_average_age'
]
# 1. Ratio between international and domestic prestige
merged_data['prestige_ratio'] = (
    merged_data['international_prestige'] / (merged_data['domestic_prestige'] + 1)
)

# 2. Experience-to-age ratio (proxy for longevity vs recency)
merged_data['experience_ratio'] = (
    merged_data['coach_experience'] / merged_data['whole_team_average_age']
)

# 3. Difference between starting XI and total team age
merged_data['age_gap'] = (
    merged_data['starting_xi_average_age'] - merged_data['whole_team_average_age']
)

# Add to your features
features += ['prestige_ratio', 'experience_ratio', 'age_gap']

target_continuous = 'overall'
target_binary = 'success_top_tier'

X = merged_data[features]
y_continuous = merged_data[target_continuous]
y_binary = merged_data[target_binary]

# Handle missing values if any
X = X.fillna(X.mean())

# Split data
X_train, X_test, y_train_cont, y_test_cont, y_train_bin, y_test_bin = train_test_split(
    X, y_continuous, y_binary, test_size=0.2, random_state=42
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nTraining data shape:", X_train.shape)
print("Test data shape:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report

# --- Random Forest Regression ---
print("\nRandom Forest Regression (predicting overall)...")
rf_reg = RandomForestRegressor(random_state=42)
rf_reg.fit(X_train_scaled, y_train_cont)
y_pred_rf = rf_reg.predict(X_test_scaled)
mse_rf = mean_squared_error(y_test_cont, y_pred_rf)
print(f"Random Forest MSE: {mse_rf:.2f}")

# --- Random Forest Classification ---
print("\nRandom Forest Classification (predicting top tier success)...")
rf_clf = RandomForestClassifier(
    random_state=42,
    max_depth=13,
    min_samples_split=55,
    n_estimators=200
)
rf_clf.fit(X_train_scaled, y_train_bin)
y_pred_rf_clf = rf_clf.predict(X_test_scaled)
acc_rf = accuracy_score(y_test_bin, y_pred_rf_clf)
print(f"Random Forest Accuracy: {acc_rf:.2f}")
print("\nClassification Report for Random Forest:")
print(classification_report(y_test_bin, y_pred_rf_clf))
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report

# --- Linear Regression ---
print("\nLinear Regression (predicting overall)...")
lr = LinearRegression()
lr.fit(X_train_scaled, y_train_cont)
y_pred_lr = lr.predict(X_test_scaled)
mse_lr = mean_squared_error(y_test_cont, y_pred_lr)
print(f"Linear Regression MSE: {mse_lr:.2f}")

# --- Logistic Regression ---
print("\nLogistic Regression (predicting top tier success)...")
logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train_bin)
y_pred_log = logreg.predict(X_test_scaled)
acc_log = accuracy_score(y_test_bin, y_pred_log)
print(f"Logistic Regression Accuracy: {acc_log:.2f}")
print("\nClassification Report for Logistic Regression:")
print(classification_report(y_test_bin, y_pred_log))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def recommend_coaches_with_model(team_need, model, n=5):
    team_need_df = pd.DataFrame([team_need])
    for feature in features:
        if feature not in team_need_df:
            team_need_df[feature] = coach_profiles_filled[feature].mean()
    team_need_df = team_need_df[features]
    team_need_scaled = scaler.transform(team_need_df)
    
    coach_profiles_scaled = scaler.transform(coach_profiles_filled)
    
    probs = model.predict_proba(coach_profiles_scaled)[:, 1]  # Probability of success
    similarities = cosine_similarity(team_need_scaled, coach_profiles_scaled)[0]
    
    recommendations = pd.DataFrame({
        'coach_id': coach_profiles_filled.index.get_level_values('coach_id'),
        'coach_name': coach_profiles_filled.index.get_level_values('short_name'),
        'success_prob': probs,
        'similarity_score': similarities
    })

    # Sort by combination of success prob + similarity
    recommendations['combined_score'] = 0.5 * recommendations['success_prob'] + 0.5 * recommendations['similarity_score']
    return recommendations.sort_values('combined_score', ascending=False).head(n)
```

```python
# Assuming you already have merged_data and features list
coach_profiles = merged_data.groupby(['coach_id', 'short_name'])[features].mean()
coach_profiles_filled = coach_profiles.fillna(coach_profiles.mean())
example_team_need = {
    'coach_experience': 10,
    'team_quality_ratio': 0.33,
    'nationality_alignment': 1,
    'international_prestige': 8,
    'domestic_prestige': 7,
    'starting_xi_average_age': 27,
    'whole_team_average_age': 25
}

print("\n🔍 Recommendations using Logistic Regression:")
print(recommend_coaches_with_model(example_team_need, logreg))

print("\n🌲 Recommendations using Random Forest:")
print(recommend_coaches_with_model(example_team_need, rf_clf))

print("\n🚀 Recommendations using XGBoost:")
print(recommend_coaches_with_model(example_team_need, xgb))
```

```python
# Get recommendations using XGBoost
top_xgb_recommendations = recommend_coaches_with_model(example_team_need, xgb)

# Get the first recommended coach
top_coach_id = top_xgb_recommendations.iloc[0]['coach_id']
top_coach_name = top_xgb_recommendations.iloc[0]['coach_name']

print(f"\n🎯 Top recommended coach by XGBoost: {top_coach_name} (ID: {int(top_coach_id)})")

# Display all attributes for this coach (use merged_data or coach_profiles)
top_coach_data = merged_data[merged_data['coach_id'] == top_coach_id]

# If multiple rows exist (e.g., for different teams), average or display a few
print("\n📋 Attributes of the top coach:")
print(top_coach_data[features + ['short_name']].drop_duplicates().head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_team_vs_coach(team_need, merged_data, coach_id):
    # Ensure coach_id column is the correct type (float or int)
    merged_data['coach_id'] = merged_data['coach_id'].astype(float)

    # Check if coach exists
    if coach_id not in merged_data['coach_id'].values:
        print(f"❌ Coach with ID {coach_id} not found in the data.")
        return

    # Get coach row
    top_coach = merged_data[merged_data['coach_id'] == coach_id].iloc[0]

    # Extract matching features
    try:
        top_coach_profile = {k: top_coach[k] for k in team_need}
    except KeyError as e:
        print(f"❌ Missing feature in dataset: {e}")
        return

    # Build comparison DataFrame
    comparison_df = pd.DataFrame({
        'Feature': list(team_need.keys()),
        'Team Need': list(team_need.values()),
        'Top Coach': list(top_coach_profile.values())
    })

    # Plotting
    fig, ax = plt.subplots(figsize=(10, 6))
    bar_width = 0.35
    index = range(len(comparison_df))

    ax.bar(index, comparison_df['Team Need'], bar_width, label='Team Need', color='skyblue')
    ax.bar([i + bar_width for i in index], comparison_df['Top Coach'], bar_width, label='Top Coach', color='orange')

    ax.set_xlabel('Features')
    ax.set_ylabel('Values')
    ax.set_title('Comparison of Team Needs vs. Top Recommended Coach (XGBoost)')
    ax.set_xticks([i + bar_width / 2 for i in index])
    ax.set_xticklabels(comparison_df['Feature'], rotation=45, ha='right')
    ax.legend()

    plt.tight_layout()
    plt.show()
plot_team_vs_coach(example_team_need, merged_data, top_coach_id)
```
