```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set up display options
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)

# Load the datasets
coaches = pd.read_csv("C:/Users/hamou/Downloads/data/male_coaches.csv")
teams = pd.read_csv("C:/Users/hamou/Downloads/data/male_teams.csv")

# Display basic info
print("Coaches dataset shape:", coaches.shape)
print("Teams dataset shape:", teams.shape)
```

```python
# Explore coaches data
print("\nCoaches data columns:")
print(coaches.columns.tolist())
print("\nMissing values in coaches data:")
print(coaches.isnull().sum())

# Explore teams data
print("\nTeams data columns:")
print(teams.columns.tolist())
print("\nMissing values in teams data:")
print(teams.isnull().sum())

# Clean coaches data
# Convert dob to age
current_year = datetime.now().year
coaches['age'] = pd.to_datetime(coaches['dob']).dt.year.apply(lambda x: current_year - x)

# Drop columns with high missing values or less relevance
coaches_clean = coaches.drop(['face_url', 'coach_url'], axis=1)

# Clean teams data
# Identify columns with all null values
null_columns = teams.columns[teams.isnull().all()]
teams_clean = teams.drop(null_columns, axis=1)

# Keep only relevant columns for our analysis
relevant_team_columns = [
    'team_id', 'team_name', 'league_name', 'league_level', 
    'nationality_name', 'overall', 'attack', 'midfield', 
    'defence', 'coach_id', 'international_prestige', 
    'domestic_prestige', 'starting_xi_average_age', 
    'whole_team_average_age'
]
teams_clean = teams_clean[relevant_team_columns]

# Display cleaned data
print("\nCleaned coaches data shape:", coaches_clean.shape)
print("Cleaned teams data shape:", teams_clean.shape)
```

```python
# Merge datasets on coach_id
merged_data = pd.merge(teams_clean, coaches_clean, on='coach_id', how='left')

# Calculate some new features
# Coach's experience (simplified as age for now)
merged_data['coach_experience'] = merged_data['age']

# Team quality metrics
merged_data['team_quality_ratio'] = merged_data['overall'] / (merged_data['attack'] + merged_data['midfield'] + merged_data['defence'])

# Nationality alignment (whether coach and team share nationality)
merged_data['nationality_alignment'] = (merged_data['nationality_name_x'] == merged_data['nationality_name_y']).astype(int)

# Display merged data
print("\nMerged data shape:", merged_data.shape)
print("Merged data columns:", merged_data.columns.tolist())
```

```python
# For this exercise, we'll define success as:
# 1. Team overall rating (continuous)
# 2. Top tier classification (binary - above/below median)

median_overall = merged_data['overall'].median()
merged_data['success_top_tier'] = (merged_data['overall'] >= median_overall).astype(int)

print("\nSuccess metric distribution:")
print(merged_data['success_top_tier'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Select features and target
features = [
    'coach_experience', 'nationality_alignment',
    'international_prestige', 'domestic_prestige',
    'starting_xi_average_age', 'whole_team_average_age'
]
# 1. Ratio between international and domestic prestige
merged_data['prestige_ratio'] = (
    merged_data['international_prestige'] / (merged_data['domestic_prestige'] + 1)
)

# 2. Experience-to-age ratio (proxy for longevity vs recency)
merged_data['experience_ratio'] = (
    merged_data['coach_experience'] / merged_data['whole_team_average_age']
)

# 3. Difference between starting XI and total team age
merged_data['age_gap'] = (
    merged_data['starting_xi_average_age'] - merged_data['whole_team_average_age']
)

# Add to your features
features += ['prestige_ratio', 'experience_ratio', 'age_gap']

target_continuous = 'overall'
target_binary = 'success_top_tier'

X = merged_data[features]
y_continuous = merged_data[target_continuous]
y_binary = merged_data[target_binary]

# Handle missing values if any
X = X.fillna(X.mean())

# Split data
X_train, X_test, y_train_cont, y_test_cont, y_train_bin, y_test_bin = train_test_split(
    X, y_continuous, y_binary, test_size=0.2, random_state=42
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nTraining data shape:", X_train.shape)
print("Test data shape:", X_test.shape)
```

```python
# ===================== 1. FEATURE ENGINEERING =====================

# 1.1 Calculate coach age
coaches['dob'] = pd.to_datetime(coaches['dob'], errors='coerce')
coaches['coach_age'] = datetime.now().year - coaches['dob'].dt.year

# 1.2 Calculate team prestige
teams['team_prestige'] = (teams['international_prestige'] + teams['domestic_prestige']) / 2

# 1.3 Calculate financial strength
teams['financial_strength'] = teams['club_worth_eur'] + teams['transfer_budget_eur']

# 1.4 Create extra features table
extra_features = teams[['coach_id', 'team_prestige', 'financial_strength', 'whole_team_average_age']].drop_duplicates('coach_id')

# 1.5 Add coach age
extra_features = extra_features.merge(
    coaches[['coach_id', 'coach_age']], 
    on='coach_id', 
    how='left'
)

# 1.6 🔥 NEW MERGED DATA
merged_for_model = merged_data.merge(extra_features, on='coach_id', how='left')

# ===================== 2. FEATURE SELECTION =====================

features = [
    'coach_experience', 
    'team_quality_ratio', 
    'nationality_alignment',
    'prestige_ratio',
    'experience_ratio',
    'age_gap',
    'team_prestige',
    'financial_strength',
    'whole_team_average_age_x'  # 👈 Corrected here
]

X = merged_for_model[features]
y = merged_for_model['success_top_tier']

# ===================== 3. DATA PREPROCESSING =====================

from sklearn.impute import SimpleImputer

X_encoded = pd.get_dummies(X, drop_first=True)
imputer = SimpleImputer(strategy='mean')
X_encoded = imputer.fit_transform(X_encoded)

# Important: reconvert to DataFrame after imputing
X_encoded = pd.DataFrame(X_encoded, columns=X.columns, index=X.index)


# ===================== 4. MODEL TRAINING =====================

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# ===================== 5. EVALUATION =====================

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

y_pred_logreg = model.predict(X_test)

print("📈 Accuracy:", accuracy_score(y_test, y_pred_logreg))
print("\n📋 Classification Report:\n", classification_report(y_test, y_pred_logreg))

cm = confusion_matrix(y_test, y_pred_logreg)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Your models (example dictionary)
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
    "XGBoost": xgb.XGBClassifier(eval_metric='logloss', use_label_encoder=False, random_state=42),
    "LightGBM": lgb.LGBMClassifier(n_estimators=100, random_state=42),
    "CatBoost": CatBoostClassifier(iterations=100, random_state=42, verbose=0)
}

# Common function to plot learning curve
def plot_learning_curve(model, title, X, y, cv=5, scoring='roc_auc', n_jobs=-1):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=cv, scoring=scoring, n_jobs=n_jobs,
        train_sizes=np.linspace(0.01, 1.0, 5), shuffle=True, random_state=42
    )

    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)

    plt.figure(figsize=(8,6))
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.title(f"Learning Curve: {title}")
    plt.xlabel("Training examples")
    plt.ylabel("ROC AUC Score")
    plt.legend(loc="best")
    plt.grid()
    plt.xlim(0, train_sizes[-1])
    plt.ylim(0, 1.05)  # <<== add this line
    plt.show()



# Now plot for each model
for name, model in models.items():
    plot_learning_curve(model, name, X_encoded, y)
```
