Data Preparation:

```python
# 3.1 Conversion des dates

# Conversion des colonnes de dates en format datetime
df['subscription_date'] = pd.to_datetime(df['subscription_date'])
df['renewal_date'] = pd.to_datetime(df['renewal_date'])
```

```python
# 3.2 Cr√©ation de la variable 'duree_abonnement'

# Calcul de la dur√©e en jours entre la souscription et le renouvellement
df['duree_abonnement'] = (df['renewal_date'] - df['subscription_date']).dt.days
```

```python
# 3.3 Encodage de la variable 'statusInsurance'

# Remplacer ACTIVE par 1, CANCELLED ou NaN par 0
df['statusInsurance_enc'] = df['statusInsurance'].map({'ACTIVE': 1, 'CANCELLED': 0})
df['statusInsurance_enc'] = df['statusInsurance_enc'].fillna(0)
```

```python
# 3.4 Encodage de la variable cat√©gorielle 'typeInsurance'

# Encodage one-hot de typeInsurance (en supprimant la premi√®re colonne pour √©viter la multicolin√©arit√©)
df = pd.get_dummies(df, columns=['typeInsurance'], drop_first=True)
```

```python
# 3.5 V√©rification des valeurs manquantes restantes

# V√©rification g√©n√©rale des colonnes
df.isnull().sum()
```

 Modeling:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 5.1 S√©lection des variables pour le clustering
variables_clustering = ['amount', 'duree_abonnement', 'statusInsurance_enc'] + \
                       [col for col in df.columns if col.startswith('typeInsurance_')]

df_clustering = df[variables_clustering]
```

```python
# 5.2 Normalisation des variables

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)
```

```python
# 5.3 R√©duction de dimension (PCA) pour visualisation

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Ajout des composantes principales au DataFrame original
df['PCA1'] = df_pca[:, 0]
df['PCA2'] = df_pca[:, 1]
```

```python
# 5.4 Clustering avec KMeans (3 clusters comme point de d√©part)

kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)
```

```python
# 5.5 Visualisation des clusters

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='cluster', palette='Set2', s=100)
plt.title('Segmentation des utilisateurs par clustering (KMeans)')
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Cr√©ation du mod√®le hi√©rarchique
agglo = AgglomerativeClustering(n_clusters=3)
df['cluster_agglo'] = agglo.fit_predict(df_scaled)
```

```python
# Visualisation des clusters - Agglomerative Clustering

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='cluster_agglo', palette='coolwarm', s=100)
plt.title('Clustering hi√©rarchique (Agglomerative)')
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

R√©sultats & Interpr√©tation des Clusters:

```python
# 6.1 Statistiques descriptives par cluster - KMeans
df.groupby('cluster')[variables_clustering].mean().T
```

```python
# 6.2 Statistiques descriptives par cluster - Agglomerative
df.groupby('cluster_agglo')[variables_clustering].mean().T
```

```python
# 6.3 Taille des clusters - KMeans
df['cluster'].value_counts()
```

```python
# 6.4 Taille des clusters - Agglomerative
df['cluster_agglo'].value_counts()
```

- **Cluster 0 (KMeans)** : Compos√© majoritairement de polices √† faible montant et courtes dur√©es ‚Äî peut repr√©senter les petits contrats courts (ex. sport amateurs).
- **Cluster 1 (KMeans)** : Profils avec polices plus longues et montants moyens ‚Äî clients r√©guliers √† fid√©liser.
- **Cluster 2 (KMeans)** : Haute prime + longue dur√©e ‚Äî peut correspondre √† des clubs professionnels assur√©s √† haut niveau de risque.

Comparaison avec Agglomerative :
- Les regroupements sont similaires/diff√©rents sur les profils extr√™mes (√† justifier selon les moyennes).


```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Silhouette score - KMeans
silhouette_kmeans = silhouette_score(df_scaled, df['cluster'])
db_index_kmeans = davies_bouldin_score(df_scaled, df['cluster'])

# Silhouette score - Agglomerative
silhouette_agglo = silhouette_score(df_scaled, df['cluster_agglo'])
db_index_agglo = davies_bouldin_score(df_scaled, df['cluster_agglo'])

# R√©sum√©
print("üîπ KMeans")
print(f"Silhouette Score : {silhouette_kmeans:.4f}")
print(f"Davies-Bouldin Index : {db_index_kmeans:.4f}\n")

print("üî∏ Agglomerative Clustering")
print(f"Silhouette Score : {silhouette_agglo:.4f}")
print(f"Davies-Bouldin Index : {db_index_agglo:.4f}")
```

### ‚úÖ Conclusion ‚Äì Mod√®le √† retenir

Le **mod√®le KMeans** est **l√©g√®rement meilleur** sur les deux m√©triques internes de clustering :

- üìè **Silhouette Score** : `0.2867` (vs. `0.2863` pour Agglomerative)  
  ‚Üí Meilleure compacit√© et s√©paration des clusters.

- üìâ **Davies-Bouldin Index** : `1.4539` (vs. `1.4618` pour Agglomerative)  
  ‚Üí Moins d‚Äôoverlap entre les groupes, meilleure qualit√© de segmentation.

üëâ **KMeans est donc retenu** pour la suite de l‚Äôanalyse et l‚Äôinterpr√©tation des profils utilisateurs.
